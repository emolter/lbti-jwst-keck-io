# Make PSF with photutils

Testing photutils.psf functionality. Are the few lucky PSF frames Al sent me sufficient to make an oversampled PSF?

Following the tutorial here: https://photutils.readthedocs.io/en/stable/epsf.html#build-epsf

```python
import numpy as np
import matplotlib.pyplot as plt
import paths
from astropy.nddata import NDData
from photutils.psf import extract_stars
from astropy.io import fits
from astropy.table import Table
from photutils.detection import find_peaks
from photutils.psf import EPSFBuilder
```

```python
hdul = fits.open(paths.data / 'slctdPsfCube.fits')
psf_frames = hdul[0].data
print(psf_frames[0].shape)
data = [NDData(frame) for frame in psf_frames]
peak_pix = [find_peaks(frame, threshold=600)[0] for frame in psf_frames]
test_peak = peak_pix[0]
stars_tbl = [Table([[peaktab['x_peak']], [peaktab['y_peak']]], names=('x', 'y')) for peaktab in peak_pix]
stars = extract_stars(data, stars_tbl, size=(23, 23))
```

```python
fig, ax = plt.subplots(3,3, figsize=(12, 12), squeeze=True)
ax = ax.ravel()
for i in range(9): #range(psf_frames.shape[0]):
    ax[i].imshow(stars[i], origin='lower', cmap='hot')
plt.savefig('diagnostic_plots/update_061523/psf_frames.png', dpi=300)
plt.show()
```

```python
epsf_builder = EPSFBuilder(oversampling=1, maxiters=3, progress_bar=False)  
epsf, fitted_stars = epsf_builder(stars)
```

```python
i0, i1 = 10, 13
psf_cut = np.mean(epsf.data[i0:i1,:], axis = 0)

fig, (ax0, ax1) = plt.subplots(1, 2, figsize = (10, 6))
ax0.imshow(epsf.data, origin='lower', cmap='hot')
ax1.plot(psf_cut)
#plt.colorbar()
plt.savefig('diagnostic_plots/update_061523/psf.png', dpi=300)
plt.show()
```

This is a photutils FittableImageModel() object, which is great because we need to fit it to the data.

## test PSF jitter

It seems that the apparent location of the Emakong feature jitters in the x-direction.

Testing here whether this is true for the 25 PSF frames.

Workflow: fit the combined PSF to each individual frame of the PSF, extract x,y position. Plot those x,y positions.

```python
from astropy.modeling.fitting import LMLSQFitter
```

```python
fitter = LMLSQFitter(calc_uncertainties=True)
y, x = np.mgrid[:psf_frames[0].shape[0], :psf_frames[0].shape[1]]

output_models = []
param_stds = []
for i, frame in enumerate(psf_frames):
    epsf.x_0 = 12
    epsf.y_0 = 12
    epsf.x_0.min = 9
    epsf.x_0.max = 15
    epsf.flux = np.sum(frame)
    output_model = fitter(epsf, x, y, frame, maxiter=2000)
    cov = fitter.fit_info['param_cov'] 
    param_stds.append(np.sqrt(np.diag(cov))) #flux, x_0, y_0
    
    fig, (ax0, ax1, ax2) = plt.subplots(1,3, figsize=(18, 6))
    ax0.imshow(frame, origin = 'lower')
    ax0.set_title('Individual PSF frame')
    ax1.imshow(output_model(x, y), origin = 'lower')
    ax1.set_title('Fit to combined PSF')
    ax2.imshow(frame - output_model(x, y), origin='lower')
    ax2.set_title('Residual')
    if i < 3:
        plt.savefig(f'diagnostic_plots/update_061523/psf_frame_fit_{i}.png', dpi=300)
    plt.show()
    
    output_models.append(output_model)
```

```python
param_stds = np.array(param_stds)
xf = np.array([m.x_0.value for m in output_models])
sigxf = np.array(param_stds[:,1])
yf = np.array([m.y_0.value for m in output_models])
sigyf = np.array(param_stds[:,2])

fig, ax = plt.subplots(1,1, figsize = (6,6))
ax.errorbar(xf, yf, xerr=sigxf, yerr=sigyf, linestyle = '')
ax.errorbar(np.mean(xf), np.mean(yf), xerr=np.std(xf), yerr=np.std(yf), marker = 'o', color = 'k', markersize=10)
ax.set_xlabel('X position')
ax.set_ylabel('Y position')
ax.set_xlim(10.5, 12.5)
ax.set_ylim(10.8, 12.8)
plt.gca().set_aspect('equal')
plt.savefig('diagnostic_plots/update_061523/psf_frame_xy.png', dpi=300)
plt.show()

print(f'X = {np.mean(xf):.2f} +/- {np.std(xf):.2f} pixels')
print(f'Y = {np.mean(yf):.2f} +/- {np.std(yf):.2f} pixels')
print(f'Average single-fit sigma X = {np.mean(sigxf):.2f}')
print(f'Average single-fit sigma Y = {np.mean(sigyf):.2f}')
```

## Make an FT of the combined PSF

```python
from scipy import fft

def to_visibilities(data):
    
    transform = fft.fftn(data)
    amp2 = transform.real**2 - transform.imag**2
    phase = np.arctan(transform.imag / transform.real)
    return amp2, phase


def calc_uvdist(data, pixscale_arcsec, wl_obs):
    '''
    wl_obs in meters to get uvdist in meters
    '''
    
    x,y = (data.shape[0]/2, data.shape[1]/2)
    xx, yy = np.mgrid[-x:x, -y:y]
    r = np.sqrt(xx**2 + yy**2)
    theta = np.deg2rad(pixscale_arcsec/3600)
    
    uvdist = (0.5 * wl_obs) / (r*theta)
    return uvdist
```

```python
pixscale_arcsec = 0.01071 #arcsec, lbti
amp2, phase = to_visibilities(epsf.data)
uvdist = calc_uvdist(epsf.data, pixscale_arcsec, 5.0e-6)

fig, (ax0, ax1, ax2) = plt.subplots(3,1, figsize = (6, 9))
    
ax0.imshow(epsf.data, origin = 'lower')
#ax0.axvline(sz/2 + 1, linestyle=':', color = 'k')
ax0.set_xlabel('Pixel')
ax0.set_ylabel('Flux')

ax1.scatter(uvdist.flatten(), np.sqrt(amp2.flatten()), marker = '.', color = 'k')
#ax1.axhline(meanflux, color = 'red', linestyle = '--', label = 'Mean Derived Emakong Flux')
ax1.set_ylabel('Amplitude')
ax1.set_yscale('log')
ax1.legend()

ax2.scatter(uvdist.flatten(), phase.flatten(), marker = '.', color = 'k')
ax2.set_ylabel('Phase')
ax2.set_xlabel('UVdist (m)')

for ax in [ax1, ax2]:
    ax.set_xlim([0,23])

fig.savefig('diagnostic_plots/update_061523/combined_psf_ft.png', dpi=300)
plt.show()
```
